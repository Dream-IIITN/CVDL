```python
# Read and Display an Image
demo = cv2.imread('dog.jpg')
plt.imshow(demo)
plt.show()
# Use cv2.imread() to load an image and cv2.imshow() to display it.
```

```python
# Draw a Circle on an Image--cv2.circle(image, center_coordinates, radius, color, thickness)
circlex=cv2.circle(demo,(50,50),3,(255, 0, 0) ,1)
plt.imshow(circlex)
plt.show()
```

```python
# Draw a Line on an Image--cv2.line(image, start_point, end_point, color, thickness)
linex = cv2.line(demo, (100,150), (125,200), (255,0,0), 3)
# Use cv2.line() to draw a straight line between two points on the image.
plt.imshow(linex)
plt.show()
```

```python
# Draw a Rectangle on an Image--cv2.rectangle(image, start_point, end_point, color, thickness)
recx = cv2.rectangle(demo, (50,50), (100,100), (255,0,0), 1)
plt.imshow(recx)
plt.show()
# Use cv2.rectangle() to draw a rectangle with specified corners.
```

```python
# Draw Multiple Shapes on an Image
x = cv2.line(demo, (150,150), (125,200), (255,0,0), 8)
x = cv2.line(demo, (50,150), (25,200), (255,0,0), 8)
plt.imshow(x)
plt.show()
# Combine different shapes like circles, lines, and rectangles on the same image.
```

```python
# Convert Image to Grayscale
fx = cv2.cvtColor(demo, cv2.COLOR_BGR2GRAY)
plt.imshow(fx, cmap='gray')
plt.show()
# Use cv2.cvtColor() to convert the image to grayscale and display it.
```

```python
# Resize an Image-Syntax: cv2.resize(source, dsize, dest, fx, fy, interpolation)
# Parameters:
# source: Input Image array (Single-channel, 8-bit or floating-point) 
# dsize: Size of the output array
# dest: Output array (Similar to the dimensions and type of Input image array) [optional]
# fx: Scale factor along the horizontal axis  [optional]
# fy: Scale factor along the vertical axis  [optional]
# interpolation: One of the above interpolation methods  [optional]
rx = cv2.resize(demo, (1050, 1610))
plt.imshow(rx)
plt.show()
# Use cv2.resize() to resize the image to a new width and height.
```

```python
# Rotate an Image
height, width = demo.shape[:2]
center = (width/2, height/2) 
rox = cv2.getRotationMatrix2D(center=center, angle=90, scale=1)
roe = cv2.warpAffine( src=demo, M=rox, dsize=(width, height)) 
plt.imshow(roe)
plt.show()
# Use cv2.getRotationMatrix2D() to rotate the image by a specified angle.
```
