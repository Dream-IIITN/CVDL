```python
import cv2
print(cv2.__version__)
```

```python
img = cv2.imread('dog.jpg')
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.imshow(img)#looking green due to BGR but default RGB --cv2
plt.show()
```

```python
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_rgb)#now correct input see difference
plt.show()
```

```python
img_rgb.shape
```

```python
type(img_rgb)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray)#why not gray?? looking green due to default color scheme
plt.show()
```

```python
plt.imshow(gray, cmap='gray')#make colormap gray---> grayscale
plt.show()
```

```python
plt.imshow(img_rgb[:,:,0], cmap='gray')#make color channel red ---> grayscale
plt.show()
```

```python
plt.imshow(img_rgb[:,:,1], cmap='gray')#make color channel green ---> grayscale
plt.show()
```

```python
plt.imshow(img_rgb[:,:,2], cmap='gray')#make color channel blue ---> grayscale
plt.show()
```

## Find the difference between above 3 Images??

```python
fig = plt.figure(figsize=(10,6))
axe = fig.add_subplot(111)
axe.imshow(img_rgb)
axe.axis('off')
axe.set_title('Image DIsplay!')
```

```python
resizable = cv2.resize(img_rgb, (300,300))
```

```python
fig = plt.figure(figsize=(10,6))
axe = fig.add_subplot(111)
axe.imshow(resizable)
axe.axis('off')
axe.set_title('Image DIsplay!')
```

```python
cv2.line(img_rgb,(10,10), (50,20),(0,0,255),5)
plt.imshow(img_rgb)
plt.show()
```
